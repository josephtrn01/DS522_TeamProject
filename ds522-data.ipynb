```python
import pandas as pd
```

```python
# Load NASA GITSTEMP temperature anomaly data
nasa_temp = pd.read_csv('GLB.Ts+dSST.csv', skiprows=1)
nasa_temp
```

```python
# Load NOAA CO2 annual mean data (skip header rows)
noaa_co2 = pd.read_csv('co2_annmean_mlo.csv', skiprows=57)
noaa_co2
```

```python
# Load EM-DAT natural disaster data
emdat = pd.read_excel('public_emdat_custom_request_2025-05-21_b9841311-29d5-4247-af0f-dee413ab4f8e.xlsx')
emdat
```

```python
# Load World Bank Indicators data
world_bank = pd.read_csv('API_19_DS2_en_csv_v2_144674.csv', skiprows=4)
world_bank
```

We are going to clean data of each datasets from organizations we picked and just only focus on the annual temperature anomaly and ensure it is numeric. Rows with bad data are dropped

```python
# Clean NASA Temperature Data
# Keep only Year and J-D (annual average anomaly)
nasa_temp_clean = nasa_temp[['Year', 'J-D']]

# Rename columns
nasa_temp_clean.columns = ['Year', 'Temp_Anomaly']

# Convert anomalies to numeric (remove "***" or bad data)
nasa_temp_clean['Temp_Anomaly'] = pd.to_numeric(nasa_temp_clean['Temp_Anomaly'], errors='coerce')

# Drop rows with missing values
nasa_temp_clean.dropna(inplace=True)
nasa_temp_clean
```

We only need the Year and CO₂ concentration and clean up bad rows for NOAA CO2 Data

```python
# Clean NOAA CO₂ Data
# Rename columns properly
noaa_co2.columns = ['Year', 'CO2_ppm', 'Uncertainty']

# Keep only Year and CO2 concentration
noaa_co2_clean = noaa_co2[['Year', 'CO2_ppm']]

# Ensure Year is integer
noaa_co2_clean['Year'] = pd.to_numeric(noaa_co2_clean['Year'], errors='coerce')

# Drop missing years
noaa_co2_clean.dropna(inplace=True)
noaa_co2_clean
```

We count how many disasters happened per Country-Year for EM-DAT Disaster Data

```python
# Clean EM-DAT Disaster Data
# Focus on Start Year, Country, Disaster Type
emdat_clean = emdat[['Start Year', 'Country', 'Disaster Type']]

# Rename Start Year to Year for consistency
emdat_clean = emdat_clean.rename(columns={'Start Year': 'Year'})

# Remove rows with missing Year or Country
emdat_clean = emdat_clean.dropna(subset=['Year', 'Country'])

# Ensure Year is integer
emdat_clean['Year'] = pd.to_numeric(emdat_clean['Year'], errors='coerce')
emdat_clean = emdat_clean.dropna(subset=['Year'])
emdat_clean['Year'] = emdat_clean['Year'].astype(int)

# Group by Year and Country, count number of disasters
disaster_counts = emdat_clean.groupby(['Year', 'Country']).size().reset_index(name='Disaster_Count')
disaster_counts
```

We focus on total population and reshape the data from wide to long for merging for World Bank Data

```python
# Clean World Bank Data
# Filter World Bank for total population
population_data = world_bank[world_bank['Indicator Name'] == 'Population, total']

# Keep only Country Code, Country Name, and year columns
population_data = population_data[['Country Name', 'Country Code'] + [str(y) for y in range(1960, 2024)]]

# Reshape from wide to long
population_long = population_data.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Population')

# Convert Year to int, Population to numeric
population_long['Year'] = pd.to_numeric(population_long['Year'], errors='coerce')
population_long['Population'] = pd.to_numeric(population_long['Population'], errors='coerce')

population_long.dropna(inplace=True)
population_long
```

```python
# Merge global-level data (Temperature Anomalies + CO2 Trends)
# Merge on Year only (since these are global, no country breakdown)
global_data = pd.merge(nasa_temp_clean, noaa_co2_clean, on='Year', how='inner')

# --- Merge country-level data (Disaster Counts + Population) ---
# First, align disaster_counts and population_long on Country and Year
disaster_pop_data = pd.merge(disaster_counts, 
                             population_long, 
                             left_on=['Country', 'Year'], 
                             right_on=['Country Name', 'Year'], 
                             how='inner')

# Calculate disasters per million people
disaster_pop_data['Disasters_per_Million'] = (disaster_pop_data['Disaster_Count'] / disaster_pop_data['Population']) * 1_000_000

# --- Combine country-level data with global trends ---
# Note: Since global data has no country dimension, we join only on Year (attach same global values to all countries)
final_data = pd.merge(disaster_pop_data, 
                      global_data, 
                      on='Year', 
                      how='left')

# Select and reorder important columns
final_data = final_data[['Year', 
                         'Country', 
                         'Country Code', 
                         'Disaster_Count', 
                         'Population', 
                         'Disasters_per_Million', 
                         'Temp_Anomaly', 
                         'CO2_ppm']]

# Preview the final merged dataset
final_data
```

```python
# Save cleaned and merged dataset
final_data.to_csv('cleaned_climate_disaster_data.csv', index=False)
```
